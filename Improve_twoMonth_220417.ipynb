```python
# 获得当天日期
import datetime

today = datetime.datetime.today()
year = str(today.year)
month = ("" if today.month > 9 else "0") + str(today.month)
day = str(today.day)

date = "" + str(year) + str(month) + str(day)

# 获得前一个月日期
oneMonthAgo = datetime.datetime.now() - datetime.timedelta(days = 62)
year_oneMonth = str(oneMonthAgo.year)
month_oneMonth = ("" if oneMonthAgo.month > 9 else "0") + str(oneMonthAgo.month)
day_oneMonth = str(oneMonthAgo.day)
date_oneMonth = "" + str(year_oneMonth) + str(month_oneMonth) + str(day_oneMonth)

print("输出从", date_oneMonth, "到", date, "间两个月的数据")
```

```python
workbook = xlrd.open_workbook("./shares.xlsx")
worksheet = workbook.sheet_by_index(0)
```

```python
col_data = worksheet.col_values(1)  #获取第一列的内容
names=worksheet.col_values(3)
markets=worksheet.col_values(5)
```

```python
colums = ['ts_code', 'trade_date', 'open', 'high', 'low', 'close', 'pct_change','change','pct_chg','name','market']
with open(date + '_twoMonth.csv', 'a+',newline='',encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(colums)
    f.close()
```

```python
for i,j,k in zip(col_data,names,markets):
    print(i,j,k)

    df = pro.query('daily', ts_code=i, start_date=date_oneMonth,fields='ts_code,trade_date,open,high,low,close,pre_close,change ,pct_chg')
    print(df)

    with open(date + '_twoMonth.csv','a+',newline='',encoding='gbk') as f:
        writer=csv.writer(f)
        for row in df.values:
            row1 = np.append(row,j)
            row2 = np.append(row1,k)
            writer.writerow(row2)
```
