```python
# 获取当天日期
today = datetime.datetime.today()
year = str(today.year)
month = ("" if today.month > 9 else "0") + str(today.month)
day = ("" if today.day > 9 else "0") + str(today.day)
date = "" + str(year) + str(month) + str(day)
```

```python
# 获得前10天日期
day_10 = 10
TenDayAgo = datetime.datetime.now() - datetime.timedelta(days = day_10)
year_TenDay = str(TenDayAgo.year)
month_TenDay = ("" if TenDayAgo.month > 9 else "0") + str(TenDayAgo.month)
day_TenDay = str(("" if TenDayAgo.day > 9 else "0") + str(TenDayAgo.day))
date_TenDay = "" + str(year_TenDay) + str(month_TenDay) + str(day_TenDay)
```

```python
# 获得前一个月日期
oneMonthAgo = datetime.datetime.now() - datetime.timedelta(days = day_before)
year_oneMonth = str(oneMonthAgo.year)
month_oneMonth = ("" if oneMonthAgo.month > 9 else "0") + str(oneMonthAgo.month)
day_oneMonth = str(("" if oneMonthAgo.day > 9 else "0") + str(oneMonthAgo.day))
date_oneMonth = "" + str(year_oneMonth) + str(month_oneMonth) + str(day_oneMonth)
print(date_oneMonth)
print("输出从", date_oneMonth, "到", date, "间",day_before, "天的数据")
```
